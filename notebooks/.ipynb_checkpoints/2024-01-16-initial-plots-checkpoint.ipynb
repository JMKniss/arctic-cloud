# Plotting SHEBA Data

```python
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as ncdf
# import xarray as xr
import pandas as pd
import seaborn as sns

# import matplotlib.cm as cm
from matplotlib.colors import ListedColormap

from matplotlib.animation import PillowWriter
```

Here is the configuration:

```python
# color palette and color map
colors = [
    '#FFFFFF',
    '#E7B5D0', '#C295C1', '#9B7BB2', '#856DAF', '#6F56A0',
    '#8DD1EE', '#70BFE6', '#58AFDA', '#4A9ECD', '#4290C1',
    '#9BCEA1', '#81C38D', '#6DBA7B', '#60B068', '#5CA55A',
    '#FAE173', '#F7CE68', '#F4B55C', '#F19F4F', '#EF8A43',
    '#EC6337', '#C2633E', '#9B6449', '#80624E', '#685E50',
    '#000000',
]

cp = sns.color_palette(colors)
my_cmap = ListedColormap(cp.as_hex())
```

```python
# file_path = 'data/2024-01-16/SHEBA/lidar/may06.cl.ncdf'

# Eureka Data
# file_path = 'data/2024-01-17/eureka/ahsrl_20060910T000001_data_SEARCH_September08.nc'

# 1 min averages from SHEBA lidar
# file_paths: list[str] = [  # 19-20 November 1997 Lidar: 1 min averages
#     '',  # 0.41 - 8.75  (This has the beginning of the 6th)
#     '',  # 8.75 - 17.08 (This is the middile of the 6th)
#     '',  # 17.08 - 24 (This is the end of the 6th)
# ]

file_paths: list[str] = [  # 6 May 1998 Lidar: 1 min averages
    'data/2024-01-17/05060025.BARO.ncdf',  # 0.41 - 8.75  (This has the beginning of the 6th)
    'data/2024-01-17/05060845.BARO.ncdf',  # 8.75 - 17.08 (This is the middile of the 6th)
    'data/2024-01-17/05061705.BARO.ncdf',  # 17.08 - 24 (This is the end of the 6th)
]
```

```python
variable_map: dict[str, str] = {'power': 'far_parallel', 'depol_ratio': 'depolarization',}
variable_names: list[str] = list(variable_map.keys())
variable_names.sort()
```

## 1) Load the data using `netCDF4`

```python
# Load all of the dataframes from the file_paths list:
datasets: list[ncdf.Dataset] = [ncdf.Dataset(fp) for fp in file_paths]
```

```python
# Join of the the datasets into single dataframes for each variable

# Make a dictionary of 
dataframes: dict[str, list[pd.DataFrame]] = {
    var_name: pd.concat(
        [
            pd.DataFrame(
                ds[variable_map[var_name]][:],
                index=ds['time'][:],
                columns=ds['range'][:],
            )
            for ds in datasets
        ]
    )
    for var_name in variable_names
}
```

```python
def _display(fig, ax, data, kind='mat', aspect='auto', xlabel='Time [hours]', ylabel='Range (km)', shrink=0.95, cbloc='right', vmin=None, vmax=None):
    if kind == 'mat':
        method = ax.matshow
    elif kind == 'im':
        method = ax.imshow
    im = method(data.T, aspect=aspect, vmin=vmin, vmax=vmax, extent=[data.index.min(), data.index.max(), data.columns.max(), data.columns.min()])
    ax.invert_yaxis()
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    
    fig.colorbar(im, ax=ax, location=cbloc, shrink=shrink)

def display(dataframes, keys, kind='mat', vmin=None, vmax=None):
    length = len(keys)
    fig, axs = plt.subplots(length, 1, layout='constrained')

    if (vmin is None) and (vmax is None):
        vmin = [None] * length
        vmax = [None] * length
        
    for i, (ax, key) in enumerate(zip(axs, keys)):
        _display(fig, ax, dataframes[key], kind=kind, vmin=vmin[i], vmax=vmax[i]) 
```

```python
display(dataframes, variable_names)
```

# 3) Replace `-999` flags with `np.nan`

```python
dataframes = {
    key: value.replace(-999, np.nan)
    for key, value in dataframes.items()
}
```

```python
display(dataframes, variable_names, vmin=[-1, None], vmax=[1, None])
```

# 4) Determine the thresholds

```python
pow_extent = [dataframes['power'].index.min(), dataframes['power'].index.max(), dataframes['power'].columns.max(), dataframes['power'].columns.min()]
depol_extent = [dataframes['depol_ratio'].index.min(), dataframes['depol_ratio'].index.max(), dataframes['depol_ratio'].columns.max(), dataframes['depol_ratio'].columns.min()]
max_elevation = 12000

power_low_thresh = 50
power_high_thresh = 65
```

```python
m = (power_high_thresh - power_low_thresh) /  max_elevation
```

```python
fig = plt.figure(figsize=(12, 7), layout='constrained')
axs = fig.subplot_mosaic([["power_img", "depol_ratio_img"],
                          ["power_line", "depol_ratio_line"],
                          ["clean_power", "clean_depol_ratio"]])

# plot power signal
axs["power_img"].set_title("Power");
im = axs["power_img"].matshow(dataframes['power'].T, extent=pow_extent, aspect='auto', cmap=my_cmap, vmin=-4, vmax=104)
axs["power_img"].xaxis.tick_bottom()
axs["power_img"].set_ylim(0, max_elevation)
axs["power_img"].set_xlabel("Time [hours]")
axs["power_img"].set_ylabel("Range (km)")
fig.colorbar(im, ax=axs["power_img"], location='right', pad=0, shrink=0.95)

# plot power field
axs["power_line"].set_title("Power Fields");
axs["power_line"].fill_betweenx([0, max_elevation], [power_low_thresh, power_high_thresh], [100, 100], color='gray', alpha=0.25)
axs["power_line"].set_xlim(0, 100)
axs["power_line"].set_ylim(0, max_elevation)
axs["power_line"].set_xlabel("Power")
axs["power_line"].set_xlabel("Range (km)")

# plot depolarization ratio signal
axs["depol_ratio_img"].set_title("Depolarization Ratio");
im = axs["depol_ratio_img"].matshow(dataframes['depol_ratio'].T, extent=depol_extent, aspect='auto', cmap=my_cmap, vmin=-0.024, vmax=0.624)
axs["depol_ratio_img"].xaxis.tick_bottom()
axs["depol_ratio_img"].set_ylim(0, max_elevation)
axs["depol_ratio_img"].set_xlabel("Time [hours]")
axs["depol_ratio_img"].set_ylabel("Range (km)")
fig.colorbar(im, ax=axs["depol_ratio_img"], location='right', pad=0, shrink=0.95)

# plot depolarization ratio field
axs["depol_ratio_line"].set_title("Depolarization Ratio Fields");
axs["depol_ratio_line"].fill_betweenx([0, max_elevation], [0, 0], [0.6, 0.6], color='gray', alpha=0.25)
axs["depol_ratio_line"].set_xlim(-1, 1)
axs["depol_ratio_line"].set_ylim(0, max_elevation)
axs["depol_ratio_line"].set_xlabel("Power")
axs["depol_ratio_line"].set_xlabel("Range (km)")

power_copy = dataframes['power'].copy(deep=True)
power_copy.iloc[:] = np.nan
depol_ratio_copy = dataframes['depol_ratio'].copy(deep=True)
depol_ratio_copy.iloc[:] = np.nan

# prepare the clean power map
axs["clean_power"].set_title("Filtered power");
im = axs["clean_power"].matshow(power_copy.T, extent=pow_extent, aspect='auto', cmap=my_cmap, vmin=-2.4, vmax=62.4)
axs["clean_power"].xaxis.tick_bottom()
axs["clean_power"].set_ylim(0, max_elevation)
axs["clean_power"].set_xlabel("Time [hours]")
axs["clean_power"].set_ylabel("Range (km)")
fig.colorbar(im, ax=axs["clean_power"], location='right', pad=0, shrink=0.95)

# prepare the clean depolarization ratio map
axs["clean_depol_ratio"].set_title("Filtered Depolarization Ratio");
im = axs["clean_depol_ratio"].matshow(depol_ratio_copy.T, extent=depol_extent, aspect='auto', cmap=my_cmap, vmin=-0.024, vmax=0.624)
axs["clean_depol_ratio"].xaxis.tick_bottom()
axs["clean_depol_ratio"].set_ylim(0, max_elevation)
axs["clean_depol_ratio"].set_xlabel("Time [hours]")
axs["clean_depol_ratio"].set_ylabel("Range (km)")
fig.colorbar(im, ax=axs["clean_depol_ratio"], location='right', pad=0, shrink=0.95)

# Sliders
top_left_slider, = axs["power_img"].plot([], [], '-k')
top_right_slider, = axs["depol_ratio_img"].plot([], [], '-k')

bottom_left_slider, = axs["clean_power"].plot([], [], '-k')
bottom_right_slider, = axs["clean_depol_ratio"].plot([], [], '-k')

# Scatter plots
middle_left_scatter = axs["power_line"].scatter([], [], color='black', s=1.5)
axs["power_line"].set_axisbelow(True)
axs["power_line"].grid()

middle_right_scatter = axs["depol_ratio_line"].scatter([], [], color='black', s=2)
axs["depol_ratio_line"].set_axisbelow(True)
axs["depol_ratio_line"].grid()

# Animation settings
metadata = dict(title="Movie", artist="Rich Inman, PhD")
writer = PillowWriter(fps=20, metadata=metadata)

# Create plot
with writer.saving(fig, "temp_01.gif", 100):
    for index, time in enumerate(dataframes['power'].index[:]):
        # if index%50:
        #     continue
        top_left_slider.set_data([time, time], [0, max_elevation])
        top_right_slider.set_data([time, time], [0, max_elevation])
        bottom_left_slider.set_data([time, time], [0, max_elevation])
        bottom_right_slider.set_data([time, time], [0, max_elevation])

        offsets = np.array([[i, j] for i, j in zip(dataframes['power'].iloc[index, :], dataframes['power'].columns)])
        middle_left_scatter.set_offsets(offsets)
        
        offsets = np.array([[i, j] for i, j in zip(dataframes['depol_ratio'].iloc[index, :], dataframes['depol_ratio'].columns)])
        middle_right_scatter.set_offsets(offsets)

        # Now find out which ones passed the power test
        power_ok = [
            True if value > m * elevation + power_low_thresh
            else False
            for value, elevation
            in zip(dataframes['power'].iloc[index, :], dataframes['power'].columns)
        ]
        depol_ratio_ok = [
            True if value > 0 and value <= 0.6
            else False
            for value, elevation
            in zip(dataframes['depol_ratio'].iloc[index, :], dataframes['depol_ratio'].columns)
        ]
        set_to_nan = [False if pow and dpr else True for pow, dpr in zip(power_ok, depol_ratio_ok)]

        clean_power_data = dataframes['power'].iloc[index, :].copy(deep=True)
        clean_power_data[set_to_nan] = np.nan
        power_copy.iloc[index, :] = clean_power_data

        clean_depol_ratio_data = dataframes['depol_ratio'].iloc[index, :].copy(deep=True)
        clean_depol_ratio_data[set_to_nan] = np.nan
        depol_ratio_copy.iloc[index, :] = clean_depol_ratio_data

        # Now update the plot
        axs["clean_power"].matshow(power_copy.T, extent=pow_extent, aspect='auto', cmap=my_cmap, vmin=-2.4, vmax=62.4)
        axs["clean_power"].xaxis.tick_bottom()
        axs["clean_depol_ratio"].matshow(depol_ratio_copy.T, extent=depol_extent, aspect='auto', cmap=my_cmap, vmin=-0.024, vmax=0.624)
        axs["clean_depol_ratio"].xaxis.tick_bottom()
        
        writer.grab_frame()
        
```
